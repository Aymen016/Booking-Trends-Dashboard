### if customer gets the same room which was reserved, then room status will be 1 else 0

```python
data['room_status'] = (data['reserved_room_type'] == data['assigned_room_type']).astype(int)
data['room_status'] = data['room_status'].map({1: 'desired', 0: 'undesired'})
```

#### if adults are 2, then categorize them into couple (1), if only signle person is living, then (2), else family (3)

```python
def categorize(row):
    if row['adults'] == 2 and row['children'] == 0 and row['babies'] == 0:
        return 'Couple'
    elif row['adults'] == 1 and row['children'] == 0 and row['babies'] == 0:
        return 'Single person'
    else:
        return 'Family'
        

# Apply the function to the dataframe
data['guest_type'] = data.apply(categorize, axis=1)
```

```python
data.head()
```

```python

# Step 1: Calculate Total Guests
data['total_guests'] = data['adults'] + data['children'] + data['babies']

# Step 3: Create a pivot table with guest type and cancellation status
pivot_table = pd.pivot_table(
    data,
    values='total_guests',  # Aggregate based on total number of guests
    index='guest_type',     # Index by guest type
    columns='is_canceled',  # Columns by whether the booking was canceled
    aggfunc='sum',          # Sum the total guests
    fill_value=0            # Replace NaNs with 0
)

# Rename columns for better clarity
pivot_table.columns = ['Not Canceled', 'Cancelled']

# Calculate the Grand Total row
grand_total = pivot_table.sum().rename('Grand Total')


# Add the Grand Total to the pivot table using concat
pivot_table_with_total = pd.concat([pivot_table, grand_total.to_frame().T])

# Format the pivot table to match the desired output structure
pivot_table_with_total['Total Guests'] = pivot_table_with_total['Not Canceled'] + pivot_table_with_total['Cancelled']

# Reorder columns to match the format in the image
pivot_table_with_total = pivot_table_with_total[['Total Guests', 'Cancelled']]

# Display the final result
print(pivot_table_with_total)



# Creating a DataFrame for the data provided
guests_data = pd.DataFrame({
    'Category': ['Couple', 'Family', 'Single person'],
    'Total Guests': [163114, 49288, 22577],
    'Cancelled': [64842, 17575, 6555]
})

# Plotting the data using a bar plot
plt.figure(figsize=(10, 6))
plt.bar(guests_data['Category'], guests_data['Total Guests'], color='skyblue', label='Total Guests')
plt.bar(guests_data['Category'], guests_data['Cancelled'], color='salmon', label='Cancelled', alpha=0.7)

plt.title("Total Guests vs. Cancelled Bookings by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Displaying the plot
plt.show()




```

```python
vals = data['hotel'].value_counts()
print(vals)

# Assuming the DataFrame is named 'data'
# Plotting the count of each unique value in the 'hotel' column
hotel_counts = data['hotel'].value_counts()

# Visualizing the counts using a bar plot
plt.figure(figsize=(8, 5))
hotel_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Count of Bookings by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Number of Bookings")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)

# Display the plot
plt.show()
```

```python
# Grouping by hotel type and counting the number of cancellations
canceled_count = data.groupby('hotel')['is_canceled'].sum().reset_index()

# Renaming the column to make it clearer
canceled_count.columns = ['Hotel Type', 'Count of Canceled Bookings']

# Displaying the resulting table
print(canceled_count)

# Creating a DataFrame based on the given data
cancel_data = pd.DataFrame({
    'Hotel Type': ['City Hotel', 'Resort Hotel'],
    'Count of Canceled Bookings': [33102, 11122]
})

# Visualizing the count of canceled bookings for each hotel type using a bar plot
plt.figure(figsize=(8, 5))
plt.bar(cancel_data['Hotel Type'], cancel_data['Count of Canceled Bookings'], color=['skyblue', 'salmon'])
plt.title("Count of Canceled Bookings by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("Count of Canceled Bookings")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)

# Display the plot
plt.show()
```
